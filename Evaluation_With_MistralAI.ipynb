## Import Necessary Packages

```python
from langchain.llms import CTransformers
from langchain.evaluation.loading import load_evaluator
```

## Configuration Parameters

```python
CONCISENESS = "conciseness"
CORRECTNESS = "correctness"
MODEL = "res/mistral-7b-openorca.Q8_0.gguf"
MODEL_TYPE = "llama"
MAX_NEW_TOKENS = 512
TEMPERATURE = 0.1
DO_SAMPLE=True
```

```python
from langchain.llms import CTransformers
config = {'max_new_tokens': 256, 'repetition_penalty': 1.1,'temperature':0.9}
llm = CTransformers(model="TheBloke/Mistral-7B-OpenOrca-GGUF",
                    model_file="mistral-7b-openorca.Q8_0.gguf",config=config)
```

```python
evaluator = load_evaluator("labeled_criteria",
                            criteria="correctness",
                            llm=llm
                            )
```

```python
question = "How to Determine the Dollar Value of the Options Grant ?"
```

```python
output = """To determine the dollar value of the options grant, you need to calculate the number of shares in the grant and multiply it by the current share price.\n\nThe formula to calculate the number of shares in the options grant is:\n(Number of Shares) = Dollar Value of Options Grant / Current Share Price\n\nOnce you have calculated the number of shares, you can multiply it by the current share price to determine the dollar value of the options grant.\n\nThere are no specific quotes supporting this answer as the information provided is a general formula for calculating the dollar value of the options grant."""
```

```python
reference = """dollar value of options grant = Employee base salary * options multiplier."""
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
from langchain.schema import output
evaluation_result = evaluator.evaluate_strings(input=question,
                                              prediction=output,
                                              reference=reference)
```
