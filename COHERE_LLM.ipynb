```python
from getpass import getpass
cohere_key = getpass()
```

```python
import cohere
co = cohere.Client(cohere_key)
response = co.generate(
  model='command-nightly',
  prompt='write a description about India',
  max_tokens=300,
  temperature=0.9,
  k=0,
  p=0.75,
  stop_sequences=[],
  return_likelihoods='NONE')
print('Prediction: {}'.format(response.generations[0].text))
```

## Generating embeddings with cohere

```python
from langchain.embeddings import CohereEmbeddings
embeddings = CohereEmbeddings(cohere_api_key=cohere_key)
text = "This is a test document."
query_result = embeddings.embed_query(text)
doc_result = embeddings.embed_documents([text])
print(doc_result)
```

## Document QnA Example

#### Convert URL to PDF

```python
import weasyprint
import os
```

```python
def create_pdf(url):
  pdf = weasyprint.HTML(url).write_pdf()
  output_dir="pdf_dir"
  if not os.path.exists(output_dir):
    os.makedirs(output_dir)
  file_path = os.path.join(output_dir,'sample.pdf')
  with open(file_path,'wb') as f:
    f.write(pdf)
```

```python
url = "https://en.wikipedia.org/wiki/India"
create_pdf(url)
```

```python
from langchain.document_loaders import TextLoader
from langchain.document_loaders import PyPDFLoader,PyPDFDirectoryLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import VectorDBQA
from langchain.llms import Cohere
import os
directory = "/content/pdf_dir"
loader = PyPDFDirectoryLoader(directory)
#loader = PyPDFLoader("/content/drive/MyDrive/ChatGPT/Shared/ChromaDB/Data/Australia - Wikipedia2.pdf")
pages = loader.load_and_split()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
docs = text_splitter.split_documents(pages)

os.environ["COHERE_API_KEY"] = cohere_key
```

## Instantiate Vector store

```python
embeddings = CohereEmbeddings()
db = Chroma.from_documents(docs, embeddings)
```

## Question Answering

```python
qa = VectorDBQA.from_chain_type(llm=Cohere(), chain_type="stuff", vectorstore=db)
query = "Tell me about the population of India."
qa.run(query)
```
